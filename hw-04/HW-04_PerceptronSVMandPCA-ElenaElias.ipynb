## Part 6. Using Principal Components (11 points)

The full model uses all 10 transformed features to predict the results and you likely found that the model is decently accurate using all 10 features, but not perfect. **Could we get the same level of accuracy (or better) using fewer features?** When datasets start to get very large and complex, applying some sort of **feature reduction** method can reduce the computational resources needed to train the model and, in some case actually improve the accuracy.

When performing feature reduction, one could simply try to identify which features seem most important and drop the ones that aren't, but performing a Principal Component Analysis (PCA) to determine the features that contribute the most to the model (through their accounted variance) can be more effective.

### 6.1 Running a Principle Component Analysis (PCA)

Since we have 10 total features to start with, let's see how well we can do with just 1 feature. Reduce the feature count to **1** principle components. 

**&#9989; Task 6.1 (3 points):**  Using `PCA()` and the associated `fit()` method, run a principle component analysis on your training features using 1 component. Transform both the test and training features using the result of your PCA. Print the `explained_variance_ratio_`.

````python
# Put your code here
features = transformed.drop(columns=['Class'])
labels = transformed["Class"]

# splitting the data
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, train_size=0.8, random_state=12
)

pca = PCA(n_components=1)
pca.fit(X_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

print("Explained variance ratio:", pca.explained_variance_ratio_)

svc_default = SVC()

svc_default.fit(X_train_pca, y_train)

y_pred = svc_default.predict(X_test_pca)

accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = conf_matrix.ravel().tolist()
print("\nTN FP FN TP:", tn, fp, fn, tp)
print(conf_matrix)

````

<font size=+3>&#9998;</font> It is 0.77, the model will perform quite well, the accuracy and explained variance are both high, that variable is doing the "heavy-lifting"

### 6.3 Repeat your analysis with more components

What if we increase the number of principle components to **4**? What happens now?

**&#9989; Task 6.3 (2 points):** Repeat your analysis from 6.1 and 6.2 using **4 components** instead. As part of your analysis, **print the total explained variance ratio as well as the sum of these values**.

````python
# Put your code here
# Put your code here
features = transformed.drop(columns=['Class'])
labels = transformed["Class"]

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, train_size=0.8, random_state=12
)

from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# === PCA with 4 components ===
pca4 = PCA(n_components=4)
pca4.fit(X_train)

# Transform the data
X_train_pca4 = pca4.transform(X_train)
X_test_pca4 = pca4.transform(X_test)

# Print variance ratios
print("Explained variance ratio for each component:", pca4.explained_variance_ratio_)
print("Total explained variance:", pca4.explained_variance_ratio_.sum())

# === Train SVC using DEFAULT parameters ===
svc_default_4 = SVC()  # default hyperparameters
svc_default_4.fit(X_train_pca4, y_train)

# Predict
y_pred_4 = svc_default_4.predict(X_test_pca4)

# Evaluation
accuracy_4 = accuracy_score(y_test, y_pred_4)
print(f"\nTest set accuracy (4 PCA components): {accuracy_4:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_4))

conf_matrix_4 = confusion_matrix(y_test, y_pred_4)
tn, fp, fn, tp = conf_matrix_4.ravel().tolist()
print("\nTN FP FN TP:", tn, fp, fn, tp)
print(conf_matrix_4)
````

&#9989; **Question 6.3 (2 point):** What is the total explained variance ratio captured by this PCA? How accurate is this model? What evidence are you using to determine that? 

<font size=+3>&#9998;</font> The PCA with 4 components captures a total explained variance of about 0.975, meaning it preserves nearly all of the original information. The model achieves an accuracy of 0.914, which is supported by the classification report showing strong precision and recall values and by the confusion matrix showing only a few misclassifications. Overall, the evidence indicates that increasing to 4 components maintains high model performance, but keeps accuracy the same.

````python
# Put your code here
from sklearn.linear_model import Perceptron
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

param_grid = {
    'penalty': ['l2', 'l1', 'elasticnet'],
    'alpha': [0.0001, 0.001, 0.01, 0.1]
}

perceptron_model = Perceptron(random_state=12)

grid_search = GridSearchCV(perceptron_model, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
print("Best hyperparameters:", best_params)

best_perceptron = grid_search.best_estimator_
y_pred = best_perceptron.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Test set accuracy: {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)

tn, fp, fn, tp = conf_matrix.ravel().tolist()
print("\nTN FP FN TP:", tn, fp, fn, tp)
````
